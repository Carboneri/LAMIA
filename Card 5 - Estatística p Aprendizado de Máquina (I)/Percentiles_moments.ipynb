```python
import numpy as np
import matplotlib.pyplot as plt #chama as bibliotecas

vals = np.random.normal(0, 0.5, 10000) # aqui, estamos gerando 10.000 números aleatórios com distribuição normal, média 0 e desvio padrão 0,5. A função np.random.normal() é usada para isso.

plt.hist(vals, 50)
plt.show()
```

```python
np.percentile(vals, 50) # so para mostrar que a mediana é o percentil 50, ou seja, metade dos valores estão abaixo e metade acima.
```

```python
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(0, 0.5, 10000) # aqui, estamos gerando 10.000 números aleatórios com distribuição normal, média 0 e desvio padrão 0,5. A função np.random.normal() é usada para isso.

plt.hist(vals, 50) 
plt.show()
```

```python
np.mean(vals) # so a media
```

```python
np.var(vals) # so a variancia
```

```python
import scipy.stats as sp
sp.skew(vals) # medida de assimetria, que indica se a distribuição é simétrica ou inclinada para um lado. Um valor positivo indica uma cauda mais longa à direita, enquanto um valor negativo indica uma cauda mais longa à esquerda.
```

```python
sp.kurtosis(vals) # medida de curtose, que indica a "pontiagudez" da distribuição. Uma curtose positiva indica uma distribuição com caudas mais pesadas e um pico mais alto, enquanto uma curtose negativa indica uma distribuição com caudas mais leves e um pico mais baixo.
```
