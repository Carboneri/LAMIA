```python
import pandas as pd
import numpy as np
```

```python
from numpy.random import randn
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5,4), index='A B C D E'.split(), columns='W X Y Z'.split()) #criando um dataframe com 5 linhas e 4 colunas, usando o metodo randn para gerar numeros aleatorios seguindo uma distribuicao normal, usando o metodo seed para garantir que os numeros aleatorios sejam os mesmos toda vez que o codigo for executado, usando o metodo DataFrame para criar o dataframe e passando os parametros de numeros aleatorios, index e columns
```

```python
df
```

```python
df > 0 #printa quais sao os numeros da tabela que sao maiores que 0
```

```python
bol = df > 0 # cria variavel bol para receber teste condicional
```

```python
df[bol] #metodo se comporta igual numpy, printa todos aqueles que sao maior que zero e nan pros q nao
```

```python
df[df['W']>0] #printa todas as linhas que contem os numeros q na coluna W sao maiores que zero
```

```python
df[df['W']>0]['Y'] #printa todas as linhas que contem os numeros q na coluna W sao maiores que zero mas limitando a coluna Y
```

```python
bol = df['W']>0 #mesma coisa que o de cima so q mais verboso
df2 = df[bol]
df2['Y']
```

```python
df[(df['W']>0) & (df['Y']>1)] #faz condicionais and mas utilizando o & se n da erro, mas retorna as linhas de W e Y q sao maiores q 0 e 1
```
