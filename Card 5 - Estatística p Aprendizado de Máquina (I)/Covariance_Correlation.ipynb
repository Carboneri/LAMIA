Covariance and correlation

```python
import numpy as np
import matplotlib.pyplot as plt

def de_mean(x): 
    xmean = np.mean(x) # calcula a media
    return [xi - xmean for xi in x] # retorna os valores sem a media


def covariance(x, y): 
    n = len(x) 
    return np.dot(de_mean(x), de_mean(y)) / (n-1) # calcula

pageSpeeds = np.random.normal(3.0, 1.0, 1000) 
purchaseAmount = np.random.normal(50.0, 10.0, 1000) 

plt.scatter(pageSpeeds, purchaseAmount)
covariance (pageSpeeds, purchaseAmount)
```

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds 
plt.scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

```python
def correlation(x, y):
    stddevx = x.std() 
    stddevy = y.std() 
    return covariance(x,y) / stddevx / stddevy #na vida real vc deveria checar se stddevx e stddevy são diferentes de zero, para evitar divisão por zero.

correlation(pageSpeeds, purchaseAmount) 
```

```python
import numpy as numpy
numpy.corrcoef(pageSpeeds, purchaseAmount)
```

```python
purchaseAmount = 100 - pageSpeeds * 3 
plt.scatter(pageSpeeds, purchaseAmount)
correlation (pageSpeeds, purchaseAmount)
```
