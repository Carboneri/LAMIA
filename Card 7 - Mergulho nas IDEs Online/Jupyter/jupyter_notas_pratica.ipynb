## Exercicio Pratico Jupyter
Na primeira prática, utilizei o Jupyter Notebook para criar um notebook local voltado à análise simples de notas de alunos. Foi gerado um gráfico de barras para visualizar melhor as notas.

```python
import pandas as pd
import matplotlib.pyplot as plt
## Criando um DataFrame com os dados dos alunos e suas notas
dados = {
    "Aluno": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo", "Fernanda"],
    "Nota": [8.5, 6.0, 4.5, 9.0, 7.0, 5.5]
}

df = pd.DataFrame(dados)
# Adicionando uma nova coluna "Situação" com base na nota
df["Situação"] = df["Nota"].apply(lambda nota: "Aprovado" if nota >= 6 else "Reprovado")
```

```python
media = df["Nota"].mean()
maior_nota = df["Nota"].max()
menor_nota = df["Nota"].min()
situacoes = df["Situação"].value_counts()
#faz media, maior e menor nota, e conta quantos alunos foram aprovados e reprovados
print("Média da turma:", media)
print("Maior nota:", maior_nota)
print("Menor nota:", menor_nota)
print("\nSituação dos alunos:")
print(situacoes)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df["Aluno"], df["Nota"])
plt.title("Notas dos Alunos")
plt.xlabel("Aluno")
plt.ylabel("Nota")
plt.ylim(0, 10)
plt.show()
# faz um gráfico de barras com as notas dos alunos, com o eixo y limitado de 0 a 10
```
