<b>Importe NumPy como np

```python
import numpy as np
```

<b>Crie uma matriz de 10 zeros

```python
np.zeros(10) #matriz de 10 zeros
```

<b>Crie uma matriz de 10 ones

```python
np.ones(10)#matriz de 10 uns
```

<b>Crie uma matriz de 10 cincos

```python
np.linspace(5, 5, 10) #vai de 5(inicio) em 5(fim) dez vezes
```

<b>Crie um array de inteiros de 10 até 50

```python
np.arange(10, 51) #vai de 10(inicio) seria indice 0 e vai ate 51(fim) que seria indice 50
```

<b>Crie um array dos numeros pares de 10 até 50

```python
np.arange(10,51, 2) #vai de 10(inicio) seria indice 0 e vai ate 51(fim) que seria indice 50, acrescentando o 2 que vai ser os intervalos de numeros
```

<b>Criei uma matriz 3x3 com valores variando de 0 até 8

```python
arr = np.arange(0,9) #crio a um array de 0 a 8 e dps dou um reshape para transformar em uma matriz de 3x3
arr.reshape((3,3))
```

<b>Crie uma matriz identidade 3x3

<b>Use NumPy para gerar números aleatórios entre 0 e 1

<b>Use Numpy para gerar um array de 25 números aleatórios tirados de uma distribuição normal.

<b >Indexação Numpy e Seleção </b><br>
Agora você receberá algumas matrizes e será solicitado a replicar as saídas resultantes da matriz:

```python
# matriz 5x5 com valores sequenciais de 1 a 25
mat = np.arange(1,26).reshape(5,5)
#chama matriz
mat
```

```python
mat[2: , 1: ] #comeca da linha indice 2 e vai da coluna indice 1 ate o final
```

```python
mat[3, 4] #pega o numero 20 
```

```python
mat[0:3, 1:2] #delimita aos numeros da linha 1 a 3 e da coluna 2
```

```python
mat[4, 0: ] #printa a linha 4 toda
```

```python
mat[3: ,0: ] #printa as duas primeiras linhas e todas as colunas
```

<b>Agora faça o seguinte</b><br>
Obter a soma de todos os valores no "mat"    


```python
mat.sum() #soma a matriz
```

<b>Obter o desvio padrão dos valores em mat

```python
mat.std() #desvio padrao da matriz com a funcao q faz isso
```

<b>Obter a soma de todas as colunas em mat

```python
mat.sum(axis=0) #soma as colunas
```
