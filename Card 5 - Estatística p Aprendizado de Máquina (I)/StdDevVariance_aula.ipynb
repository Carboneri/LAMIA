<h1> <b> Standard Deviation and Variance </h1> 

```python
import numpy as np
import matplotlib.pyplot as plt
incomes = np.random.normal(100.0, 20.0, 10000)

plt.hist(incomes, 50)
plt.show() #Gerar um histograma dos dados de renda usando a função hist() do Matplotlib, com 50 bins, para visualizar a distribuição dos valores de renda no conjunto de dados.
```

```python
incomes.std() #Calcular o desvio padrão dos dados de renda usando a função std() do NumPy para medir a dispersão dos valores de renda em relação à média.
```

```python
incomes.var() #Calcular a variância dos dados de renda usando a função var() do NumPy para medir a dispersão dos valores de renda em relação à média, e entender a relação entre variância e desvio padrão, onde a variância é o quadrado do desvio padrão.
```
