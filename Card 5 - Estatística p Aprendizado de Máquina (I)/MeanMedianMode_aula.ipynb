<h2><b>Mean vs. Median</h2>

vamos criar alguns dados falsos de renda, centrados em torno de 27.000 com uma distribuição normal e desvio padrão de 15.000, com 10.000 pontos de dados. (vamos discutir esses termos mais tarde, se você não estiver familiarizado com eles.) 

então, calcule a média - deve estar próxima de 27.000

```python
import numpy as np

incomes = np.random.normal(27000, 15000, 10000) #Gerar uma distribuição normal de rendas com média 27000 e desvio padrão 15000, usando a função de geração de números aleatórios do NumPy.
np.mean(incomes) 
```

We can segment the income data into 50 buckets, and plot it as a histogram:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(incomes, 50) 
plt.show() #Gerar um histograma com 50 bins para os dados de renda, usando a função de geração de números aleatórios do NumPy para criar uma distribuição normal com média 27000 e desvio padrão 15000.
```

Now compute the median - since we have a nice, even distribution it too should be close to 27.000

```python
np.median(incomes) #Calcular a mediana dos dados de renda usando a função de geração de números aleatórios do NumPy para criar uma distribuição normal com média 27000 e desvio padrão 15000.
```

Now we'll add Jeff Bezos into de mix. Darn income inequality

```python
incomes = np.append(incomes, [1000000000]) #Adicionar um valor extremo (outlier) de 1 bilhão aos dados de renda usando a função de geração de números aleatórios do NumPy para criar uma distribuição normal com média 27000 e desvio padrão 15000.
```

The median won't change much, but the mean does:

```python
np.median(incomes) #Calcular a mediana dos dados de renda usando a função de geração de números aleatórios do NumPy para criar uma distribuição normal com média 27000 e desvio padrão 15000.
```

```python
np.mean(incomes) #Calcular a média dos dados de renda usando a função de geração de números aleatórios do NumPy para criar uma distribuição normal com média 27000 e desvio padrão 15000, e adicionando um valor extremo (outlier) de 1 bilhão.
```

<h2> Mode

Next, let's generate some fake age data for 500 people

```python
ages = np.random.randint(18,high=90,size=500) #Gerar uma distribuição de idades com valores inteiros entre 18 e 90, usando a função de geração de números aleatórios do NumPy para criar uma distribuição uniforme com 500 amostras.
ages
```

```python
from scipy import stats
stats.mode(ages) #Calcular a moda dos dados de idade usando a função de geração de números aleatórios do NumPy para criar uma distribuição uniforme com valores inteiros entre 18 e 90, e usando a função de moda do SciPy para encontrar o valor mais frequente.
```
