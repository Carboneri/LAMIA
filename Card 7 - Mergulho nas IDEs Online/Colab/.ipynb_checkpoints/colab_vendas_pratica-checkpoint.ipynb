```python
import pandas as pd
import matplotlib.pyplot as plt

dados = {
    "Produto": ["Notebook", "Mouse", "Teclado", "Monitor", "Mouse", "Notebook", "Teclado"],
    "Quantidade": [2, 5, 3, 1, 4, 1, 2],
    "Valor_Unitario": [3500, 80, 150, 900, 80, 3500, 150]
}

df = pd.DataFrame(dados)

df["Total"] = df["Quantidade"] * df["Valor_Unitario"]

df
```

```python
total_vendido = df["Total"].sum()
media_venda = df["Total"].mean()
produto_mais_vendido = df.groupby("Produto")["Quantidade"].sum().idxmax()

print("Total vendido:", total_vendido)
print("Média de venda:", media_venda)
print("Produto mais vendido:", produto_mais_vendido)
```

```python
vendas_por_produto = df.groupby("Produto")["Total"].sum()

vendas_por_produto.plot(kind="bar", figsize=(8, 5), title="Total vendido por produto")
plt.xlabel("Produto")
plt.ylabel("Total vendido")
plt.show()
```
