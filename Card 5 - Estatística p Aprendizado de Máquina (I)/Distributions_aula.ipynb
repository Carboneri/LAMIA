<h1><b>Exemples of Data Distributions</h1>

<h2><b>Uniform Distribution</h2> 

```python
import numpy as np
import matplotlib.pyplot as plt
# Gerar 100.000 números aleatórios entre -10 e 10
values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

<h2><b>Normal / Gaussian</h2>

Visualize the probability density function:

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
# Gerar um gráfico da função de densidade de probabilidade da distribuição normal padrão
x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

Generate some random numbers with a normal distribution. "mu" is the desired mean, "sigma" is the standard deviation:

```python
import numpy as np
import matplotlib.pyplot as plt
# Gerar 10.000 números aleatórios de uma distribuição normal com média 5 e desvio padrão 2
mu = 5.0
sigma = 2.0
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

<h2><b>Exponential PDF / "Power Law"</h2>

```python
from scipy.stats import expon
import matplotlib.pyplot as plt
# Gerar um gráfico da função de densidade de probabilidade da distribuição exponencial com lambda = 1
x = np.arange(0,10,0.001)
plt.plot(x, expon.pdf(x))
```

<h2><b>Binomial Probability Mass Function</h2>

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
# Gerar um gráfico da função de massa de probabilidade da distribuição binomial com n = 10 e p = 0.5
n, p = 10, 0.5
x = np.arange(0, 10, 0.001) 
plt.plot(x, binom.pmf(x,n,p))
```

<h2><b>Poisson Probability Mass Function</h2>

Exemple: My website gets on average 500 visits per day. What's the odds of getting 550?

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
# Gerar um gráfico da função de massa de probabilidade da distribuição de Poisson com lambda = 500
mu = 500
x = np.arange(400,600,0.5)
plt.plot(x, poisson.pmf(x, mu))
```
