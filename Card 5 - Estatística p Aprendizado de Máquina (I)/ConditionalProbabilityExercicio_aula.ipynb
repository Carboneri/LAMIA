<h1><b>Condicional Probability Activity & Exercise</h1>

```python
from numpy import random
random.seed(0) #setar uma seed para garantir que os resultados sejam reproduzíveis. 
#Criamos um dicionário para contar o número total de pessoas em cada década de idade e outro 
#para contar o número de compras feitas por pessoas em cada década de idade. Em seguida, simulamos 
#100.000 pessoas, escolhendo aleatoriamente uma década de idade para cada pessoa e determinando se ela fez uma 
#compra com base na probabilidade associada à sua década de idade. Por fim, imprimimos os resultados.

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0} 
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0 
for _ in range(100000): 
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    purchaseProbability = float(ageDecade) / 100.0 
    totals[ageDecade] += 1 
    if (random.random() < purchaseProbability): 
        totalPurchases += 1 
        purchases[ageDecade] += 1 
```

```python
totals # Imprime o número total de pessoas em cada década de idade
```

```python
purchases # Imprime o número de compras feitas por pessoas em cada década de idade
```

```python
totalPurchases # Imprime o número total de compras feitas
```

```python
PEF = float(purchases[30]) / float(totals[30]) # Calcula a probabilidade de compra para pessoas na década de 30 anos
print('P(purchase | 30s): ' + str(PEF))
```

```python
PF = float(totals[30]) / 100000.0 # Calcula a probabilidade de uma pessoa ser da década de 30 anos
print("P(30's): "+ str(PF))
```

```python
PE = float(totalPurchases) / 100000.0 # Calcula a probabilidade de uma pessoa fazer uma compra, independentemente da década de idade
print("P(Purchase): " + str(PE))
```

```python
print("P(30's, purchase)"+str(float(purchases[30])/ 100000.0)) # Calcula a probabilidade de uma pessoa ser da década de 30 anos e fazer uma compra
```

```python
print("P(30's)P(Purchase)"+str(PE * PF)) # Calcula a probabilidade de uma pessoa ser da década de 30 anos e fazer uma compra, assumindo que as duas variáveis são independentes
```

```python
print((purchases[30] / 100000.0) / PF) # Calcula a probabilidade de uma pessoa fazer uma compra, dado que ela é da década de 30 anos, usando a fórmula da probabilidade condicional: P(purchase | 30's) = P(30's, purchase) / P(30's)
```

Modifique o código acima para que a probabilidade de compra NÃO varie com a idade, tornando E e F realmente independentes.

Em seguida, confirme que P(E|F) é aproximadamente igual a P(E), mostrando que a probabilidade condicional de compra para uma determinada idade não é diferente da probabilidade a priori de compra, independentemente da idade.

```python
from numpy import random
random.seed(0) #mesma coisa que antes, mas com uma probabilidade fixa de compra, igual para todas as idades.

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0} 
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0 
for _ in range(100000):
    ageDecade = random.choice([20, 30, 40, 50, 60, 70])
    
    # Probabilidade fixa de compra, igual para todas as idades
    purchaseProbability = 0.4
    
    totals[ageDecade] += 1
    
    if random.random() < purchaseProbability:
        totalPurchases += 1
        purchases[ageDecade] += 1
```

```python
PEF = float(purchases[30]) / float(totals[30]) # Calcula a probabilidade de compra para pessoas na década de 30 anos
print("P(purchase | 30s): " + str(PEF))
```

```python
PE = float(totalPurchases) / 100000.0 # Calcula a probabilidade de uma pessoa fazer uma compra, independentemente da década de idade
print("P(Purchase): " + str(PE))
```
