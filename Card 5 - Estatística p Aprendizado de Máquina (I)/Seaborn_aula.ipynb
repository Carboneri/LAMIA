<h1><b>Seaborn</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv") #Carregar um conjunto de dados de eficiência de combustível usando a função de leitura de CSV do Pandas, a partir de uma URL fornecida.
gear_counts = df['# Gears'].value_counts()
gear_counts.plot(kind='bar')
```

```python
import seaborn as sns
sns.set() #Configurar o estilo dos gráficos do Seaborn para melhorar a aparência visual dos gráficos gerados.
```

```python
gear_counts.plot(kind='bar') #Gerar um gráfico de barras para mostrar a contagem de cada número de marchas usando a função de plotagem do Pandas, e configurando o estilo dos gráficos do Seaborn para melhorar a aparência visual dos gráficos gerados.
```

```python
df.head() #Exibir as primeiras linhas do conjunto de dados usando a função head() do Pandas para obter uma visão geral dos dados de eficiência de combustível carregados.
```

```python
sns.distplot(df['CombMPG']) #Gerar um gráfico de distribuição para a coluna 'CombMPG' do conjunto de dados usando a função distplot() do Seaborn, para visualizar a distribuição dos valores de milhas por galão combinados (CombMPG) dos veículos no conjunto de dados.
```

```python
df2 = df[['Cylinders','CityMPG','HwyMPG','CombMPG']]
df2.head() #Criar um novo DataFrame df2 contendo apenas as colunas 'Cylinders', 'CityMPG', 'HwyMPG' e 'CombMPG' do conjunto de dados original, e exibir as primeiras linhas do novo DataFrame usando a função head() do Pandas para verificar os dados selecionados.
```

```python
sns.pairplot(df2, hue='Cylinders', height=2.5); #Gerar um gráfico de dispersão para cada par de variáveis numéricas no DataFrame df2 usando a função pairplot() do Seaborn, e colorir os pontos com base na variável 'Cylinders' para visualizar as relações entre as variáveis de eficiência de combustível e o número de cilindros dos veículos no conjunto de dados.
```

```python
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df) #Gerar um gráfico de dispersão usando a função scatterplot() do Seaborn para mostrar a relação entre a variável 'Eng Displ' (deslocamento do motor) e 'CombMPG' (milhas por galão combinados) no conjunto de dados, para visualizar como o deslocamento do motor afeta a eficiência de combustível dos veículos.
```

```python
sns.jointplot(x="Eng Displ", y="CombMPG", data=df) #Gerar um gráfico de dispersão conjunto usando a função jointplot() do Seaborn para mostrar a relação entre a variável 'Eng Displ' (deslocamento do motor) e 'CombMPG' (milhas por galão combinados) no conjunto de dados, para visualizar como o deslocamento do motor afeta a eficiência de combustível dos veículos, e incluir histogramas marginais para cada variável para mostrar suas distribuições individuais.
```

```python
sns.lmplot(x="Eng Displ", y="CombMPG", data=df) #Gerar um gráfico de regressão linear usando a função lmplot() do Seaborn para mostrar a relação entre a variável 'Eng Displ' (deslocamento do motor) e 'CombMPG' (milhas por galão combinados) no conjunto de dados, para visualizar como o deslocamento do motor afeta a eficiência de combustível dos veículos, e incluir uma linha de regressão para mostrar a tendência geral dos dados.
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.boxplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45) #Configurar o tamanho da figura usando sns.set() para definir as dimensões do gráfico, e gerar um gráfico de caixa usando a função boxplot() do Seaborn para mostrar a distribuição dos valores de 'CombMPG' (milhas por galão combinados) para cada fabricante de veículos ('Mfr Name') no conjunto de dados, e rotacionar os rótulos do eixo x em 45 graus para melhorar a legibilidade dos nomes dos fabricantes.
```

```python
ax=sns.swarmplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45) #Gerar um gráfico de enxame usando a função swarmplot() do Seaborn para mostrar a distribuição dos valores de 'CombMPG' (milhas por galão combinados) para cada fabricante de veículos ('Mfr Name') no conjunto de dados, e rotacionar os rótulos do eixo x em 45 graus para melhorar a legibilidade dos nomes dos fabricantes.
```

```python
ax=sns.countplot(x='Mfr Name', data=df) 
ax.set_xticklabels(ax.get_xticklabels(),rotation=45) #Gerar um gráfico de contagem usando a função countplot() do Seaborn para mostrar a contagem de veículos para cada fabricante ('Mfr Name') no conjunto de dados, e rotacionar os rótulos do eixo x em 45 graus para melhorar a legibilidade dos nomes dos fabricantes.
```

```python
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean')
sns.heatmap(df2) #Criar uma tabela dinâmica usando a função pivot_table() do Pandas para calcular a média de 'CombMPG' (milhas por galão combinados) para cada combinação de 'Cylinders' (número de cilindros) e 'Eng Displ' (deslocamento do motor), e gerar um mapa de calor usando a função heatmap() do Seaborn para visualizar as médias de eficiência de combustível para diferentes combinações de número de cilindros e deslocamento do motor.
```
