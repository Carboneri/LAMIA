Substitua todos os valores "NA" de salario pela media dos salarios

```python
emp['salario'].fillna(value=emp['salario'].mean()) #preencher os valores faltantes da coluna salario com a média dos salários
```

Faça outra versão preenchendo os salários ausentes com a média salarial de cada departamento.

```python
emp[emp["salario"] == emp["salario"].fillna(
    emp.groupby("departamento")["salario"].transform("mean")
)]['salario'] #preencher os valores faltantes da coluna salario com a média dos salários do departamento correspondente
```

```python
sla = emp.groupby('departamento')['salario'].mean()
sla #calcular a média salarial por departamento e exibir o resultado
```
