```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://www.kaggle.com/code/gabrielcarboneri/kaggle-pratica pratica publico no kaggle
df = pd.read_csv("/kaggle/input/competitions/titanic/train.csv") # Lê o arquivo CSV e cria um DataFrame do dataset do kaggle

df.head() # pega as primeiras linhas do dataset
```

```python
print("Linhas e colunas:", df.shape)

print("\nValores ausentes por coluna:")
print(df.isnull().sum())
# mostra o número de valores ausentes em cada coluna do DataFrame
```

```python
sobreviventes = df["Survived"].value_counts()

print("0 = Não sobreviveu")
print("1 = Sobreviveu")
print(sobreviventes)
#printa o número de sobreviventes e não sobreviventes no dataset.
```

```python
taxa_por_sexo = df.groupby("Sex")["Survived"].mean()
# Calcula a taxa de sobrevivência por sexo, agrupando os dados pelo sexo e calculando a média da coluna "Survived" para cada grupo.
print("Taxa de sobrevivência por sexo:")
print(taxa_por_sexo)
```

```python
sns.countplot(data=df, x="Sex", hue="Survived")
plt.title("Sobrevivência por Sexo no Titanic")
plt.xlabel("Sexo")
plt.ylabel("Quantidade")
plt.show()
#grafico de barras mostrando a quantidade de sobreviventes e não sobreviventes por sexo.
```
