##carregando bibliotecas necessarias

```python
import requests
from bs4 import BeautifulSoup as bs #importantado as bibliotecas do video
```

##Carregando nossa primeira pagina

```python
#Aqui é carregada a primeira página do exemplo. Depois o conteúdo é convertido para BeautifulSoup para conseguir visualizar e navegar pelo HTML.
#LCarregando conteudo da pagina web
r = requests.get("https://keithgalli.github.io/web-scraping/example.html")

#convertendo para um objeto beautiful soap
soup = bs(r.content)

#printando o html

print(soup.prettify())
```

##usando beautiful soup para Scrape

find e find_all #funcoes da biblioteca

```python
#usadondo o find para pegar apenas o primeiro elemento h2 encontrado na página.
first_header = soup.find("h2")
first_header
```

```python
# find_all para pegar todos os elementos h2 da página.
headers = soup.find_all("h2")
headers
```

```python
# o find recebe uma lista de tags, então ele procura o primeiro h1 ou h2 que aparecer.
#Passando uma lista de elementos para procurar
first_header = soup.find(["h1","h2"])
first_header
```

```python
# find_all procura todos os h1 e h2 da página, não apenas o primeiro.
headers = soup.find_all(["h1", "h2"])
headers
```

```python
# Aqui é buscado um parágrafo comum, mostrando como pegar elementos HTML pela tag.
#podemos passar atributos para o find/find_all function
paragraph = soup.find("p")
paragraph
```

```python
# Aqui o parágrafo é buscado usando um atributo específico, no caso o id. Isso deixa a busca mais precisa.

paragraph = soup.find("p", attrs={"id": "paragraph-id"})
paragraph
```

```python
# Aqui a busca é feita de forma encadeada, entrando no body, depois na div e depois no h1.
# podemos aninhar find/find_all
body = soup.find("body")
div = body.find("div")
header = div.find("h1")
header
```

```python
# a busca procura parágrafos que tenham exatamente uma string específica dentro deles.
# podemos procurar strings especificas em nossas find/find_all
paragraph = soup.find_all("p", string="Some bold text")
paragraph
```

```python
# é usado regex para procurar textos que combinam com um padrão, deixando a busca mais flexível.
import re #essa biblioteca ajuda a achar nesse exemplo a string exata que queremos
paragraph = soup.find_all("p", string=re.compile("Some"))

paragraph
```

```python
#  o regex também é usado para encontrar headers que tenham 'Header' ou 'header'.
headers = soup.find_all("h2", string=re.compile("(H|h)eader"))
headers
```

### select (CSS selector)

```python

print(soup.body.prettify())
```

```python
#usando selector CSS para pegar parágrafos que estão dentro de uma div.
content = soup.select("div p")
content
```

```python
# o selector CSS busca parágrafos que aparecem depois de um h2, mostrando outra forma de navegar no HTML.
paragraph = soup.select("h2 ~ p")
paragraph
```

```python
# a busca é mais específica, pegando o texto em negrito dentro do parágrafo que tem o id indicado.
bold_text = soup.select("p#paragraph-id b")
bold_text
```

```python
#  Aqui são selecionados parágrafos filhos diretos do body e depois são buscados elementos i dentro deles.
paragraphs = soup.select("body > p")
print(paragraphs)

for paragraph in paragraphs:
    print(paragraph.select("i"))
```

```python

#pegando por elemento com propriedades especificas
soup.select("[align=middle]")
```

Grab all fun facts that use word "is"

```python
# os fatos curiosos que possuem a palavra 'is', usando selector CSS e regex.
a = soup.select("ul.fun-facts li")
facts = [fact.find(string=re.compile("is")) for fact in a]
facts = [fact.find_parent().get_text() for fact in facts if fact]
facts
```

Download an image

```python
# feito o download de uma imagem da página, montando a URL completa e salvando o arquivo localmente.
images = soup.select("div.row div.column img")
image_url = images[0]['src']
full_url = "https://keithgalli.github.io/web-scraping/" + image_url

img_data = requests.get(full_url).content
with open('lake_como.jpg', 'wb') as handler:
    handler.write(img_data)
```

Solve the mystery challenge

```python
#  é resolvido o desafio final, acessando vários arquivos da página e procurando a informação escondida.
files = soup.select("div.block a")
relative_files = [f['href'] for f in files]

url = "https://keithgalli.github.io/web-scraping/"
for f in relative_files:
  full_url = url + f
  page = requests.get(full_url)
  bs_page = bs(page.content)
  secret_word_element = bs_page.find("p", attrs={"id": "secret-word"})
  secret_word = secret_word_element.string
  print(secret_word)
```
