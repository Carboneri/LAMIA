# Questão prática — Web Scraping em um site de livros

## Enunciado

Vou criar um notebook para acessar o site Books to Scrape e coletar informações dos livros da primeira página. O objetivo é extrair o título, preço, disponibilidade e avaliação de cada livro, organizar tudo em um DataFrame e salvar o resultado em um arquivo CSV.

```python
# Aqui eu importo as bibliotecas que vou usar para acessar o site, ler o HTML e organizar os dados.
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Aqui eu faço a requisição da página e transformo o HTML em um objeto BeautifulSoup.
url = "https://books.toscrape.com/"

resposta = requests.get(url)
soup = BeautifulSoup(resposta.content, "html.parser")

print("Status da requisição:", resposta.status_code)
```

```python
# Aqui eu encontro todos os livros da página.
livros = soup.select("article.product_pod")

print("Quantidade de livros encontrados:", len(livros))
```

```python
# Aqui eu percorro cada livro e extraio as informações principais.
dados_livros = []

for livro in livros:
    titulo = livro.select_one("h3 a")["title"]
    preco = livro.select_one(".price_color").get_text(strip=True)
    disponibilidade = livro.select_one(".availability").get_text(" ", strip=True)
    avaliacao = livro.select_one(".star-rating")["class"][1]

    dados_livros.append({
        "titulo": titulo,
        "preco": preco,
        "disponibilidade": disponibilidade,
        "avaliacao": avaliacao
    })
```

```python
# Aqui eu transformo os dados coletados em um DataFrame para visualizar melhor.
df_livros = pd.DataFrame(dados_livros)

df_livros.head()
```

```python
# Aqui eu faço uma conferência simples dos dados coletados.
print("Linhas e colunas:", df_livros.shape)

print("\nQuantidade de livros por avaliação:")
print(df_livros["avaliacao"].value_counts())
```

```python
# Aqui eu salvo o resultado em CSV para poder usar depois em outras análises.
df_livros.to_csv("livros_extraidos.csv", index=False)

print("Arquivo livros_extraidos.csv salvo com sucesso!")
```
